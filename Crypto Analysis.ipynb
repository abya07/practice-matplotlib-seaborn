```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crypto_prices_df = pd.read_csv('crypto_daily_prices.csv') 
print(crypto_prices_df)
```

```python
#Using Matplotlib, plot lineplots that display crypto daily prices on one single figure.
crypto_prices_df.plot(x = 'Date', y = ['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price'], label = ['Bitcoin', 'Ethereum', 'LiteCoin'],  linewidth = 3, figsize = (14, 6));
plt.title('Crypto Prices')
plt.ylabel("Prices[$]")
plt.legend(loc = "upper left")
plt.grid()  
plt.show()
```

```python
#Using Matplotlib, plot crypto daily prices on multiple subplots.
crypto_prices_df.plot(x = 'Date', title = 'Crypto Prices', subplots = True, grid = True, figsize = (15, 25))
plt.show() 
```

```python
crypto_returns_df = pd.read_csv('crypto_daily_returns.csv')
print(crypto_returns_df)
```

```python
#Using Matplotlib, plot the scatterplot between Bitcoin and Ethereum daily returns.
X = crypto_returns_df['BTC']
Y = crypto_returns_df['ETH']
plt.figure(figsize = (7,7))
plt.scatter(X, Y)
plt.grid()
plt.show()
```

```python
#Using Seaborn, plot similar scatterplot between bitcoin and Ethereum daily returns.
plt.figure(figsize = (7,7))
sns.scatterplot(x = 'BTC', y = 'ETH', data = crypto_returns_df)
plt.grid()
plt.show()
```

```python
#Assume that you became bullish on XRP and decided to allocate 60% of your assets in it. You also decided to equally divide the rest of your assets in other coins (BTC, LTC, ADA, and ETH). Using Matplotlib, plot a pie chart that shows these allocations. Use ‘explode’ attribute to increase the separation between BTC and the rest of the portfolio.
values = [60, 10, 5, 5, 10]
colors = ['r', 'g', 'y', 'b', 'm']
labels = ['XRP', 'BTC', 'LTC', 'ADA', 'ETH']
explode = [0, 0.2, 0, 0, 0]

plt.figure(figsize = (7, 7))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Allocation of Crypto')
plt.show()
```

```python
#Using Matplotlib, plot the histogram for ETH returns using 30 bins with red color.

mu = crypto_returns_df['ETH'].mean()
sigma = crypto_returns_df['ETH'].std()
num_bins = 30
plt.figure(figsize = (14,7))
plt.hist(crypto_returns_df['ETH'], num_bins, facecolor = 'red')
plt.title('Histogram: mu = ' + str(mu) + ' sigma = ' + str(sigma))
plt.grid()
plt.show()
```
